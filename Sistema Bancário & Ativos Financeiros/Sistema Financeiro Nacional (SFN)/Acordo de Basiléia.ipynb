## ✳️ O que são os Acordos de Basileia?

Os **Acordos de Basileia** são recomendações internacionais criadas pelo **Comitê de Supervisão Bancária de Basileia (BCBS)** para fortalecer a regulação e supervisão do setor bancário. São amplamente adotados pelos bancos centrais para garantir **solidez financeira e gestão adequada de riscos**.

- [Link com o índice de basileia de cada banco no site do BACEN](https://www3.bcb.gov.br/ifdata/)

---

## 📜 Basileia I (1988) – Foco em Risco de Crédito

- Introdução do conceito de **capital mínimo de 8% sobre os RWA (Ativos Ponderados pelo Risco)**.
- Utilização de **pesos fixos** para diferentes ativos, sem granularidade de risco.
- Limitação: não capturava diferenças reais de risco entre tomadores.

---

## 📐 Basileia II (2004) – Três Pilares e Modelos Internos

### 📌 Pilar 1 – Requisitos mínimos de capital
- Riscos cobertos: **crédito, mercado e operacional**
- Abordagens:
  - **PAD**: Abordagem Padrão (pesos regulatórios)
  - **IRB**: Internal Ratings-Based (modelos internos)

### 📌 Pilar 2 – Supervisão Bancária
- Avaliação da **adequação de capital** além dos requisitos mínimos.
- Introdução do **ICAAP** (Internal Capital Adequacy Assessment Process).

### 📌 Pilar 3 – Disciplina de Mercado
- Incentivo à **transparência** nas divulgações.
- Permite que o mercado discipline instituições com base em suas exposições e práticas de risco.

---

## 🛡️ Basileia III (2010–2017) – Resiliência Pós-Crise

Criado após a crise de 2008 para tornar o sistema mais **resiliente**.

### 🔒 Capital
- Ênfase em **capital de qualidade (CET1)**.
- Introdução de **buffers de capital**:
  - Conservação
  - Contracíclico
  - SIBs (instituições sistêmicas)

### 💧 Liquidez
- **LCR (Liquidity Coverage Ratio)** – cobertura de saídas de caixa em 30 dias.
- **NSFR (Net Stable Funding Ratio)** – liquidez de longo prazo.

### 📏 Alavancagem
- **Índice de alavancagem** para limitar exposição total.

## 📊 Indicadores Derivados

| Indicador           | Fórmula Básica                         | Objetivo                                  |
|--------------------|----------------------------------------|-------------------------------------------|
| Índice de Basileia | PR / RWA                               | Avaliar suficiência de capital            |
| PR                 | Soma do capital nível I e II           | Capital disponível para perdas            |
| RWA                | Ativos ponderados por risco            | Base de cálculo do capital regulatório    |
| LCR                | Ativos líquidos / Saídas em 30 dias    | Liquidez de curto prazo                   |
| NSFR               | Fontes estáveis / Necessidades estáveis| Liquidez de longo prazo                   |

## 🇧🇷 Adoção no Brasil

O Banco Central do Brasil implementa Basileia via resoluções como:

- **Res. CMN 4.557** – Estrutura de gerenciamento de riscos e capital  
- **Res. CMN 4.193** – Requisitos mínimos de PR  
- **Res. CMN 4.280** – ICAAP
