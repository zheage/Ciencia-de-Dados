# üß† SparkSession ‚Äì A interface unificada moderna

Introduzido no Spark 2.0, o `SparkSession` √© a interface de alto n√≠vel que unifica:

- `SQLContext`

- `HiveContext`

- e o pr√≥prio `SparkContext`

Ou seja, √© o ponto de entrada principal para trabalhar com DataFrames, SQL e recursos estruturados.

Internamente, o `SparkSession` cont√©m um `SparkContext` acess√≠vel via `.sparkContext`.

```python
from pyspark.sql import SparkSession

print("Iniciando SparkSession...")

spark = SparkSession.builder \
    .appName("SparkTeste") \
    .master("local[*]") \
    .getOrCreate()

print("SparkSession criado.")
```

```python
rdd = spark.sparkContext.parallelize([1,2,3,4,5,6])
print(rdd.count())
```

# üß© Comparativo R√°pido

| Item               | `SparkContext`                       | `SparkSession`                                        |
| ------------------ | ------------------------------------ | ----------------------------------------------------- |
| N√≠vel              | Baixo n√≠vel (RDDs)                   | Alto n√≠vel (DataFrames, SQL)                          |
| Abstra√ß√£o de dados | RDDs                                 | DataFrames (baseados em RDDs)                         |
| Introduzido em     | Spark 0.x                            | Spark 2.0                                             |
| Substitui          | ‚Äî                                    | `SQLContext`, `HiveContext`, encapsula `SparkContext` |
| Estilo             | Funcional                            | Declarativo e expressivo (tipo pandas)                |
| Recomendado para   | Opera√ß√µes com RDD, pipelines manuais | Trabalhar com tabelas, arquivos, MLlib, Spark SQL     |

```python
from faker import Faker
import random

fake = Faker('pt_BR')

# Gerar dados de faturas fict√≠cias
faturas = []
for _ in range(20):
    faturas.append({
        "cliente_id": fake.unique.random_int(min=1000, max=9999),
        "nome_cliente": fake.name(),
        "data_fatura": fake.date_between(start_date='-1y', end_date='today').strftime('%Y-%m-%d'),
        "valor": round(random.uniform(100, 5000), 2),
        "status": random.choice(["Paga", "Em aberto", "Atrasada"])
    })

# Criar DataFrame Spark
df_faturas = spark.createDataFrame(faturas)
```
