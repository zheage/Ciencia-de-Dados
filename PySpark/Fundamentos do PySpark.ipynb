# ðŸ§  SparkSession â€“ A interface unificada moderna

Introduzido no Spark 2.0, o `SparkSession` Ã© a interface de alto nÃ­vel que unifica:

- `SQLContext`

- `HiveContext`

- e o prÃ³prio `SparkContext`

Ou seja, Ã© o ponto de entrada principal para trabalhar com DataFrames, SQL e recursos estruturados.

Internamente, o `SparkSession` contÃ©m um `SparkContext` acessÃ­vel via `.sparkContext`.

```python
from pyspark.sql import SparkSession

print("Iniciando SparkSession...")

spark = SparkSession.builder \
    .appName("SparkTeste") \
    .master("local[*]") \
    .getOrCreate()

print("SparkSession criado.")
```

```python
rdd = spark.sparkContext.parallelize([1,2,3,4,5,6])
print(rdd.count())
```
